```python
import util
import wave
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams["figure.figsize"] = [16, 9]
```

```python
#normal ECG signals A00001, A00002, A00003, A00006
ecg_n1 = util.load('A00001')
ecg_n2 = util.load('A00002')
ecg_n3 = util.load('A00003')
ecg_n4 = util.load('A00006')

#first 1000 samples
ecg_n1s = ecg_n1[0:1000]
ecg_n2s = ecg_n2[0:1000]
ecg_n3s = ecg_n3[0:1000]
ecg_n4s = ecg_n4[0:1000]
```

```python
#plot normal ECG signals
plt.subplot(2,2,1)
wave.plotWave(ecg_n1s, 'Normal A00001', 'Index n * 0.003')
plt.subplot(2,2,2)
wave.plotWave(ecg_n2s, 'Normal A00002', 'Index n * 0.003')
plt.subplot(2,2,3)
wave.plotWave(ecg_n3s, 'Normal A00003', 'Index n * 0.003')
plt.subplot(2,2,4)
wave.plotWave(ecg_n4s, 'Normal A00006', 'Index n * 0.003')
```
