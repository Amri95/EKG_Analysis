```python
import util
import wave
import numpy as np
import matplotlib.pyplot as plt
import pywt
%matplotlib inline
plt.rcParams["figure.figsize"] = [16, 9]
```

```python
## Load data and plot
#Normal A00001
ecg_n1 = util.load('A00001')
ecg_n1s = ecg_n1[0:1000]
#others A00005
ecg_o1 = util.load('A00005')
ecg_o1s = ecg_o1[0:1000]
#AF A00004
ecg_a1 = util.load('A00004')
ecg_a1s = ecg_a1[0:1000]
#Noisy A00205
ecg_ns1 = util.load('A00205')
ecg_ns1s = ecg_ns1[0:1000]

##Plot the data
plt.subplot(2,2,1)
wave.plotWave(ecg_ns1s, 'Noisy A00205', 'Index n * 0.003')
plt.subplot(2,2,2)
wave.plotWave(ecg_n1s, 'Normal A00001', 'Index n * 0.003')
plt.subplot(2,2,3)
wave.plotWave(ecg_a1s, 'AF A00004', 'Index n * 0.003')
plt.subplot(2,2,4)
wave.plotWave(ecg_o1s, 'Others A00005', 'Index n * 0.003')
```

```python
rebuilt_normal = wave.waveletDecomp(ecg_n1, 'sym4', levels=5, mode='symmetric', omissions=([1],False))
rebuilt_noisy = wave.waveletDecomp(ecg_ns1, 'sym4', levels=5, mode='symmetric', omissions=([1],False))
```

```python
coeff_noisy = pywt.wavedecn(ecg_ns1, 'sym4', level=5, mode='symmetric')
coeff_normal = pywt.wavedecn(ecg_n1, 'sym4', level=5, mode='symmetric')
```

```python
plt.subplot(6,1,1)
util.multiplot([coeff_noisy[0][0:200], coeff_normal[0][0:200]],['noisy cA', 'normal cA'])
plt.subplot(6,1,2)
util.multiplot([coeff_noisy[1]['d'][0:200], coeff_normal[1]['d'][0:200]],['noisy cD1', 'normal cD1'])
plt.subplot(6,1,3)
util.multiplot([coeff_noisy[2]['d'][0:200], coeff_normal[2]['d'][0:200]],['noisy cD2', 'normal cD2'])
plt.subplot(6,1,4)
util.multiplot([coeff_noisy[3]['d'][0:200], coeff_normal[3]['d'][0:200]],['noisy cD3', 'normal cD3'])
plt.subplot(6,1,5)
util.multiplot([coeff_noisy[4]['d'][0:200], coeff_normal[4]['d'][0:200]],['noisy cD4', 'normal cD4'])
plt.subplot(6,1,6)
util.multiplot([coeff_noisy[5]['d'][0:200], coeff_normal[5]['d'][0:200]],['noisy cD5', 'normal cD5'])
```

```python
coeff_noisy_3l = pywt.wavedecn(ecg_ns1, 'sym4', level=3, mode='symmetric')
coeff_normal_3l = pywt.wavedecn(ecg_n1, 'sym4', level=3, mode='symmetric')
```

```python
plt.subplot(6,1,1)
util.multiplot([coeff_noisy_3l[0], coeff_normal_3l[0]],['noisy D1', 'normal D1'])
plt.subplot(6,1,2)
util.multiplot([coeff_noisy_3l[1]['d'], coeff_normal_3l[1]['d']],['noisy cD2', 'normal cD2'])
plt.subplot(6,1,3)
util.multiplot([coeff_noisy_3l[2]['d'], coeff_normal_3l[2]['d']],['noisy cD3', 'normal cD3'])
plt.subplot(6,1,4)
util.multiplot([coeff_noisy_3l[3]['d'], coeff_normal_3l[3]['d']],['noisy cD4', 'normal cD4'])
```
